```python
matriz_1=np.genfromtxt('/kaggle/input/heart-disease-uab/heart.csv',delimiter=',', skip_header=1)
matriz_2 = matriz_1[:,[3,4,5,7]]
print(matriz_2)
```

```python
matriz_2m=np.mean(matriz_2, 0)
matriz_2sd=np.sqrt(np.var(matriz_2,0))
print(matriz_2m)
print(matriz_2sd)
```

```python
matriz_normalizada= (matriz_2 - matriz_2m)/matriz_2sd
print(np.mean(matriz_normalizada, 0))
print(np.var(matriz_normalizada, 0))
```

```python
matrizM=matriz_normalizada.T.dot(matriz_normalizada)
print(matrizM)
```

```python
matrizCov=matrizM/matrizM[0,0]
print(matrizCov)
```

```python
valores,vectores= np.linalg.eig(matrizCov)
print(valores)
print(vectores.T)
```

```python
vec_val_list=[[valores[i], vectores.T[i]] for i in range(4)]
vec_val_list.sort(reverse=True)
print(vec_val_list)
```

```python
cambio_matriz=np.array([vec_val_list[i][1] for i in range(4)]).T
nueva_matriz=matriz_normalizada.dot(cambio_matriz)
print(nueva_matriz)
```

```python
valores_2=list(100*valores/valores.sum())
valores_2.sort(reverse=True)
print(valores_2)
```
